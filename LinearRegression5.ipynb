### Creating Features

```python
# lets setup an empty dataframe for our features
features = pd.DataFrame()
```

```python
# lets see the distribution of our outcome variable Property Crime
sns.distplot(df['Property\ncrime'],bins=100)
```

```python
# creating population suared
features['population_squared'] = df['Population'] * df['Population']
```

```python
# Here we are turning our murder column into 0 for rows with no instances and 1 for rows with nay instances of murder
features['Murder'] = np.where((df['Murder and\nnonnegligent\nmanslaughter'] > 0), 1, 0)
```

```python
# lets do the same for robbery setting values to 0 or 1
features['Robbery'] = np.where((df['Robbery'] > 0), 1, 0)
```

```python
# importing our target to our features df
features['PropertyCrime'] = df['Property\ncrime']
```

```python
# transfering our population over to our features df
features['population'] = df['Population']
```

```python
# Adding Larceny to our features
features['larceny'] = df['Larceny-\ntheft']
```

```python
# our current features
features.head()
```

```python
# feature selection
# importing statsmodels for p-values
import statsmodels.formula.api as smf
from statsmodels.sandbox.regression.predstd import wls_prediction_std

# setting up formula
linear_formula = 'PropertyCrime ~ population_squared+Murder+Robbery+population+larceny'

# fitting model
lm = smf.ols(formula=linear_formula, data=features).fit()


print('\n Coeff: \n', lm.params)
print('\n P-Values: \n', lm.pvalues)
print('\n nR-squared:', lm.rsquared)
```

Checking the p-values for our features, we see the p-value for our Population squared feature and our Murder feature are greater than .05. This suggest that there is probably no effect for that feature, and that dropping it would not affect the  R-squared score.

```python
# feature selection
# importing statsmodels for p-values
import statsmodels.formula.api as smf
from statsmodels.sandbox.regression.predstd import wls_prediction_std

# setting up formula
linear_formula = 'PropertyCrime ~ Robbery+population+larceny'

# fitting model
lm = smf.ols(formula=linear_formula, data=features).fit()


print('\n Coeff: \n', lm.params)
print('\n P-Values: \n', lm.pvalues)
print('\n nR-squared:', lm.rsquared)
```

Running the test again we can see that dropping the population squared feature and the murder feature has onlying lost .135% of our modeling data.

### Fitting Linear Model

```python
# grabbing all our imports
from sklearn.cross_validation import KFold
from sklearn.model_selection import cross_validate
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
from matplotlib.mlab import PCA as mlabPCA
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA 
from sklearn.feature_selection import RFE
```

```python
# fitting our X with the factors we want to use to predict property crime
X = features[['population', 'Robbery','larceny']]
```

```python
# fitting our target variable property crime
y = features['PropertyCrime']
```

```python
# here we our splitting our columns from above into a training and testing set
# using 30%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
# setting our model to var
lm = LinearRegression()
```

```python
# fitting our new model with our training data
model = lm.fit(X_train,y_train)
```

```python
model.score(X_test, y_test)
```

```python
cross_validate(model, X, y, cv=5)
```

```python
cross_val_score(lm.fit(X_train,y_train), X, y,cv= 5)
```

### Model Analysis

After fitting, training, and cross validating, our model we can see that over all we have been able to make a very accurate model for this data set.  Through 5 test, the model ranged from 97-99% accurate, while being projected to be 98% accurate. However, it maybe overfit for this specific data set, and be far less accurate on a new set of data. 

### Applying PCA

```python
feat_pca = features[['population', 'Robbery','larceny']]
```

```python
ss = StandardScaler().fit_transform(X_train)
```

```python
sklearn_pca = PCA(n_components=2)
Y_sklearn = sklearn_pca.fit_transform(ss)

print(
    'The percentage of total variance in the dataset explained by each',
    'component from Sklearn PCA.\n',
    sklearn_pca.explained_variance_ratio_
)
```

```python
model2 = lm.fit(Y_sklearn, y_train)
```

```python
model2.score(Y_sklearn, y_train)
```

```python
cross_validate(model2, X, y,cv= 5)
```

### Visualizing Predicitions

```python
CVpredictions = cross_val_predict(lm.fit(Y_sklearn, y_train), X, y, cv=5)
plt.scatter(y, CVpredictions)
plt.xlim(0,3400)
plt.ylim(0,3400)
plt.title('Predictions Vs Real Values')
plt.xlabel('Real Values')
plt.ylabel('Predicted Value')
plt.show()
```

Using this scatter to show the preditions vs the actual values.  We see that the scatter plot almost looks like a straight line at the bottom, this is a very sign.  At the top there is a little more error but still very good overall.

```python
# another scatter to see how our predictions looked
plt.scatter(CVpredictions,y-CVpredictions )
plt.xlabel('Predicted')
plt.ylabel('Residual')
plt.axhline(y=0)
plt.title('Residual vs. Predicted')
plt.show()
```

Taking another look at the predictions, we can see that our model is very accurate predicting values less than 1500 and starts to lose accuracy as the y value increases
