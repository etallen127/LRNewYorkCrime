### Creating Features

```python
# lets setup an empty dataframe for our features
features = pd.DataFrame()
```

```python
# lets see the distribution of our outcome variable Property Crime
sns.distplot(df['Property\ncrime'],bins=100)
```

```python
# creating population suared
features['population_squared'] = df['Population'] * df['Population']
```

```python
# Here we are turning our murder column into 0 for rows with no instances and 1 for rows with nay instances of murder
features['Murder'] = np.where((df['Murder and\nnonnegligent\nmanslaughter'] > 0), 1, 0)
```

```python
# lets do the same for robbery setting values to 0 or 1
features['Robbery'] = np.where((df['Robbery'] > 0), 1, 0)
```

```python
# importing our target to our features df
features['PropertyCrime'] = df['Property\ncrime']
```

```python
# transfering our population over to our features df
features['population'] = df['Population']
```

```python
# Adding Larceny to our features
features['larceny'] = df['Larceny-\ntheft']
```

```python
# our current features
features.head()
```

```python
# feature selection
# importing statsmodels for p-values
import statsmodels.formula.api as smf
from statsmodels.sandbox.regression.predstd import wls_prediction_std

# setting up formula
linear_formula = 'PropertyCrime ~ population_squared+Murder+Robbery+population+larceny'

# fitting model
lm = smf.ols(formula=linear_formula, data=features).fit()


print('\n Coeff: \n', lm.params)
print('\n P-Values: \n', lm.pvalues)
print('\n nR-squared:', lm.rsquared)
```

Checking the p-values for our features, we see the p-value for our Population squared feature and our Murder feature are greater than .05. This suggest that there is probably no effect for that feature, and that dropping it would not affect the  R-squared score.

### Fitting Linear Model

```python
# grabbing all our imports
from sklearn.cross_validation import KFold
from sklearn.model_selection import cross_validate
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
from matplotlib.mlab import PCA as mlabPCA
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA 
from sklearn.feature_selection import RFE
```

```python
# fitting our X with the factors we want to use to predict property crime
X = features[['population', 'Robbery','larceny']]
```

```python
# fitting our target variable property crime
y = features['PropertyCrime']
```

```python
# here we our splitting our columns from above into a training and testing set
# using 30%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
# setting our model to var
lm = LinearRegression()
```

```python
# fitting our new model with our training data
model = lm.fit(X_train,y_train)
```

```python
model.score(X_test, y_test)
```

```python
cross_validate(model, X, y, cv=5)
```

```python
cross_val_score(lm.fit(X_train,y_train), X, y,cv= 5)
```
