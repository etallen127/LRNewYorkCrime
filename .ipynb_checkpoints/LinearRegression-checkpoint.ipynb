# Crime Data In New York State in 2013

### Importing Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn import linear_model
from sklearn.preprocessing import OneHotEncoder
```

```python
data_path = 'https://raw.githubusercontent.com/Thinkful-Ed/data-201-resources/master/New_York_offenses/NEW_YORK-Offenses_Known_to_Law_Enforcement_by_City_2013%20-%2013tbl8ny.csv'
```

```python
# import data and setting header from the 4 row
df = pd.read_csv(data_path, header=4 )
df.head()
```

### Data Cleaning

```python
# creating visualization of null values
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
plt.show()
```

```python
df.columns
```

```python
# dropping null columns
df.drop(['Rape\n(revised\ndefinition)1','Arson3'], inplace=True, axis=1)
```

```python
# creating visualization of null values
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
plt.show()
```

```python
# checking last row
(df[-2:-1])
```

```python
# dropping the row
df = df.iloc[:348]
```

```python
# creating visualization of null values
# look like all the nulls
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
plt.show()
```

### Using apply/lambda to remove comma and convert cols to numerical columns

```python
df['Population'] = df['Population'].apply(lambda x: int(x.replace(',', '')))
```

```python
df['Violent\ncrime'] = df['Violent\ncrime'].apply(lambda x: int(x.replace(',', '')))
```

```python
df['Property\ncrime'] = df['Property\ncrime'].apply(lambda x: int(x.replace(',', '')))
```

```python
df['Larceny-\ntheft'] = df['Larceny-\ntheft'].apply(lambda x: int(x.replace(',', '')))
```

```python
df['Rape\n(legacy\ndefinition)2'] = df['Rape\n(legacy\ndefinition)2'].apply(lambda x: int(x.replace(',', '')))
```

```python
df['Robbery'] = df['Robbery'].apply(lambda x: int(x.replace(',', '')))
```

```python
df['Aggravated\nassault'] = df['Aggravated\nassault'].apply(lambda x: int(x.replace(',', '')))
```

```python
df['Burglary'] = df['Burglary'].apply(lambda x: int(x.replace(',', '')))
```

```python
df['Motor\nvehicle\ntheft'] = df['Motor\nvehicle\ntheft'].apply(lambda x: int(x.replace(',', '')))
```

```python
df.head()
```

### Visualizing Data

Can see that there is a an input that is skewing the data

```python
# Lets look up a max value to see where it is coming from
df['Population'].max()
```

```python
# looks like New York is our outlier
df[df['Population'] > 8300000]
```

```python
# dropped the New York Row
df.drop(df.index[216], inplace=True)
```

```python
# creating a grid to see how columns interact with one another and their distribution
gd = sns.PairGrid(df.dropna(), diag_sharey=False)
gd.map_upper(plt.scatter, alpha=.5)
gd.map_lower(sns.regplot, scatter_kws=dict(alpha=0))
gd.map_diag(sns.kdeplot, lw=3)
```

### Creating Features

```python
features = pd.DataFrame()
```

```python
features['population_squared'] = df['Population'] * df['Population']
```

```python
features['Murder'] = np.where((df['Murder and\nnonnegligent\nmanslaughter'] > 0), 1, 0)
```

```python
features['Robbery'] = np.where((df['Robbery'] > 0), 1, 0)
```
